# Ocorrência Acre

Neste notebook será feito uma análise mais detalhado no Estado do Acre nos anos de 2015 até 2021


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
acre = pd.read_excel('indicadoressegurancapublicauf.xlsx')
acre = acre[(acre['UF']=='Acre')&(acre['Ano']!=2022)] # filtro pra selecionar o Estado do Acre e diferente do ano de 2022
acre.drop('UF',axis=1,inplace=True) # Excluir a coluna do Estado
acre.head()
```

```python
acre.groupby('Ano').sum()
```

```python
plt.figure(figsize=(10,5))
plt.title('Número de Caso Ocorido no Acre de 2015 a 2021',fontsize=14)
sns.barplot(x=acre.groupby('Ano').sum().index,y='Ocorrências',data=acre.groupby('Ano').sum(),
            color='royalblue')
```

```python
plt.figure(figsize=(12,5))
plt.title('Ocorrências no Estado do Acre no Ano de 2015',fontsize=14)
sns.barplot(y='Tipo Crime',x='Ocorrências',data=acre[acre['Ano']==2015],
            color='royalblue')
plt.xlabel('Ocorrẽncias')
plt.ylabel('Tipo de Ocorrência');
```

As maiores ocorrências registradas aqui tem a ver com homicidio doloso, em torno de 15 ocorrências.

```python
plt.figure(figsize=(12,5))
plt.title('Ocorrências no Estado do Acre no Ano de 2016',fontsize=14)
sns.barplot(y='Tipo Crime',x='Ocorrências',data=acre[acre['Ano']==2016],
            color='royalblue')
plt.xlabel('Ocorrẽncias')
plt.ylabel('Tipo de Ocorrência');
```

Já aqui temos que o crime de homicidio doloso praticamente dobra (aproximadamente) em número de ocorrências.

```python
plt.figure(figsize=(12,5))
plt.title('Ocorrências no Estado do Acre no Ano de 2017',fontsize=14)
sns.barplot(y='Tipo Crime',x='Ocorrências',data=acre[acre['Ano']==2017],
            color='royalblue')
plt.xlabel('Ocorrẽncias')
plt.ylabel('Tipo de Ocorrência');
```

Em 2017 temos um número de caso maior em roubo de veiculos, e uma diminuição nos casos de homicidio doloso para em torno de 20, e um aumento no futo de veiculos.

```python
plt.figure(figsize=(12,5))
plt.title('Ocorrências no Estado do Acre no Ano de 2018',fontsize=14)
sns.barplot(y='Tipo Crime',x='Ocorrências',data=acre[acre['Ano']==2018],
            color='royalblue')
plt.xlabel('Ocorrẽncias')
plt.ylabel('Tipo de Ocorrência');
```

Em 2017 um aumento no roubo e furto de veiculos, aumento no casos de estrupos, e parece que o homicidio doloso teve um pequeno aumento e considerar com o ano anterior.

```python
plt.figure(figsize=(12,5))
plt.title('Ocorrências no Estado do Acre no Ano de 2019',fontsize=14)
sns.barplot(y='Tipo Crime',x='Ocorrências',data=acre[acre['Ano']==2019],
            color='royalblue')
plt.xlabel('Ocorrẽncias')
plt.ylabel('Tipo de Ocorrência');
```

Aumento no roubo e uma diminuição no furto,um aumento no roubo.

```python
plt.figure(figsize=(12,5))
plt.title('Ocorrências no Estado do Acre no Ano de 2020',fontsize=14)
sns.barplot(y='Tipo Crime',x='Ocorrências',data=acre[acre['Ano']==2020],
            color='royalblue')
plt.xlabel('Ocorrẽncias')
plt.ylabel('Tipo de Ocorrência');
```

Diminuição nos crimes de furto eroubo de veiculos.

```python
plt.figure(figsize=(12,5))
plt.title('Ocorrências no Estado do Acre no Ano de 2021',fontsize=14)
sns.barplot(y='Tipo Crime',x='Ocorrências',data=acre[acre['Ano']==2021],
            color='royalblue')
plt.xlabel('Ocorrẽncias')
plt.ylabel('Tipo de Ocorrência');
```

aumento nos crimes de roubo e furto de  veiculos, além de um aumento nos casos de estupro.

```python
px.bar(acre,x='Mês',y='Ocorrências',color=acre['Ano'].astype(str),
       barmode = 'group',labels={'color':'Ano'},
       title='Número de Ocorrências por Mês no Estado do Acres nos Anos de 2015 a 2021')
```

```python
px.bar(acre,x='Tipo Crime',y='Ocorrências',barmode='group',color=acre['Ano'].astype(str),
       title='Número de Ocorrências por Tipo no Estado do Acre nos Anos de 2015 a 2021',
       labels={'color':'Ano'})
```

```python
plt.figure(figsize=(15,5))
plt.title('Ocorrência por Tipo de Crime no Estado do Acre nos anos de 2015 a 2021',fontsize=14)
sns.barplot(x='Tipo Crime',y='Ocorrências',data=acre,hue='Ano')
plt.xlabel('Tipo de Ocorrência')
plt.xticks(rotation=70);
```

```python
plt.figure(figsize=(18,8))
plt.title('Número de Ocorrências por Mês no Estado do Acre nos anos de 2015 a 2021',
          fontsize=14)
sns.barplot(x='Mês',y='Ocorrências',data=acre,hue='Ano',ci = None)
plt.legend(loc='upper right');
```
