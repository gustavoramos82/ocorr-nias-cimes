# Análise de Ocorrências por Região

Neste notebook será feto uma análise de ocorrências por região

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
oco = pd.read_excel('/content/drive/MyDrive/Datasets/indicadoressegurancapublicauf.xlsx')
oco = oco[oco['Ano']!=2022] # tirando os dados referentes ao ano de 2022 pois só tem informações até junho.
oco.head()
```

```python
oco['UF'].unique()
```

```python
Estados = list(oco['UF'])
regiao = []
for i in Estados:
  if i == 'Acre' or  i == 'Amapá' or  i =='Amazonas' or  i == 'Pará' or  i == 'Rondônia' or  i == 'Roraima' or  i == 'Tocantins':
    regiao.append('Norte')
  elif i == 'Maranhão' or  i == 'Piauí' or  i == 'Ceará' or i == 'Bahia' or i == 'Pernambuco' or i == 'Rio Grande do Norte':
    regiao.append('Nordeste')
  elif i == 'Sergipe' or i == 'Alagoas' or i == 'Paraíba':
    regiao.append('Nordeste')
  elif i == 'Distrito Federal' or i == 'Mato Grosso do Sul' or i == 'Mato Grosso' or i == 'Goiás':
    regiao.append('Centro Oeste')
  elif i == 'Espírito Santo' or i == 'Minas Gerais' or i == 'Rio de Janeiro' or i == 'São Paulo':
    regiao.append('Sudeste')
  elif i == 'Santa Catarina' or i == 'Rio Grande do Sul' or i == 'Paraná':
    regiao.append('Sul')
```

```python
oco['Regiao'] = regiao
oco.head()
```

```python
plt.figure(figsize=(15,6))
plt.title('Número de Ocorrências por Região Durante os Anos de 2015 a 2021',fontsize=14)
sns.barplot(x='Regiao',y='Ocorrências',data=oco,hue='Ano')
plt.xlabel('Região');
```
