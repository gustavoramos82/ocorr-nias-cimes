# Análise Estado do Amapá

Neste notebook será feita a análise do Estado do Amapá

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
amapa = pd.read_excel('/content/drive/MyDrive/Datasets/indicadoressegurancapublicauf.xlsx')
amapa = amapa[amapa['Ano']!=2022] # retirando o ano de 2022
amapa = amapa[amapa['UF']=='Amapá'] # selecionando o Estado do Amapá
amapa.drop('UF',axis=1,inplace=True) # retirando a coluna UF
amapa.head()
```

```python
plt.figure(figsize=(10,5))
plt.title('Número de Ocorrências no Estado do Amapá de 2015 a 2021',
          fontsize=14)
sns.barplot(x='Ano',y='Ocorrências',data=amapa,color='royalblue');
```

```python
plt.figure(figsize=(15,6))
plt.title('Tipos de Ocorrências no Amapá nos Anos de 2015 a 2021',
          fontsize=14)
sns.barplot(x='Tipo Crime',y='Ocorrências',data=amapa,hue='Ano',ci=None)
plt.xticks(rotation=80);
```

```python
plt.figure(figsize=(18,6))
plt.title('Número de Ocorrência por Mês no Estado do Amapá nos Anos de 2015 a 2021',
          fontsize=14)
sns.barplot(x='Mês',y='Ocorrências',data=amapa,hue='Ano',ci=None)
plt.legend(loc='upper center');
```
