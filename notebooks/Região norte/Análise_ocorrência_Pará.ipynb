# Análise Estado do Pará

Neste notebook será feita a análise do Estado do Pará

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
para = pd.read_excel('/content/drive/MyDrive/Datasets/indicadoressegurancapublicauf.xlsx')
para = para[para['Ano']!=2022]
para = para[para['UF']=='Pará']
para.drop('UF',axis=1,inplace=True)
para.head()
```

```python
plt.figure(figsize=(10,6))
plt.title('Número de Ocorrências no Estado do Pará de 2015 a 2021',
          fontsize=14)
sns.barplot(x='Ano',y='Ocorrências',data=para,color='royalblue')
```

```python
plt.figure(figsize=(15,6))
plt.title('Tipos de Ocorrências no Pará nos Anos de 2015 a 2021',
          fontsize=14)
sns.barplot(x='Tipo Crime',y='Ocorrências',data=para,hue='Ano',ci=None)
plt.xticks(rotation=80);
```

```python
plt.figure(figsize=(18,6))
plt.title('Número de Ocorrência por Mês no Estado do Pará nos Anos de 2015 a 2021',
          fontsize=14)
sns.barplot(x='Mês',y='Ocorrências',data=para,hue='Ano',ci=None)
plt.legend(loc='upper center');
```
