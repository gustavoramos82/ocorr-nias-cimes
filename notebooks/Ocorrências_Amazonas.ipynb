# Ocorrências no Estado do Amazonas

Neste notebook será feito uma análise mais detalhado no Estado do Amazonas nos anos de 2015 até 2021

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
amazonas = pd.read_excel('/content/drive/MyDrive/Datasets/indicadoressegurancapublicauf.xlsx')
amazonas = amazonas[(amazonas['UF']=='Amazonas')&(amazonas['Ano']!=2022)]
amazonas.drop('UF',axis=1,inplace=True)
amazonas.head()
```

```python
amazonas.groupby('Ano').sum()
```

```python
plt.figure(figsize=(10,5))
plt.title('Número de Caso Ocorido no Amazonas de 2015 a 2021',fontsize=14)
sns.barplot(x=amazonas.groupby('Ano').sum().index,y='Ocorrências',data=amazonas.groupby('Ano').sum(),
            color='royalblue');
```

```python
plt.figure(figsize=(15,5))
plt.title('Ocorrência por Tipo de Crime no Estado do Amazonas nos anos de 2015 a 2021',fontsize=14)
sns.barplot(x='Tipo Crime',y='Ocorrências',data=amazonas,hue='Ano')
plt.xlabel('Tipo de Ocorrência')
plt.xticks(rotation=70);
```

```python
plt.figure(figsize=(18,8))
plt.title('Número de Ocorrências por Mês no Estado do Amazonas nos anos de 2015 a 2021',
          fontsize=14)
sns.barplot(x='Mês',y='Ocorrências',data=amazonas,hue='Ano',ci = None)
plt.legend(loc='upper right');
```

```python
plt.figure(figsize=(12,5))
plt.title('Número de Ocorrências de Roubo de Veiculos de 2015 a 2021 no Estado do Amazonas',fontsize=14)
sns.barplot(x='Ano',y='Ocorrências',data=amazonas[amazonas['Tipo Crime']=='Roubo de veículo'],
            color='royalblue');
```

```python
plt.figure(figsize=(12,5))
plt.title('Número de Ocorrências de Furto de veículo de 2015 a 2021 no Estado do Amazonas',fontsize=14)
sns.barplot(x='Ano',y='Ocorrências',data=amazonas[amazonas['Tipo Crime']=='Furto de veículo'],
            color='royalblue');
```

```python
plt.figure(figsize=(12,5))
plt.title('Número de Ocorrências de Estupro de 2015 a 2021 no Estado do Amazonas',fontsize=14)
sns.barplot(x='Ano',y='Ocorrências',data=amazonas[amazonas['Tipo Crime']=='Estupro'],
            color='royalblue');
```
